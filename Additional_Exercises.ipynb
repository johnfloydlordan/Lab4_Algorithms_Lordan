```python
# 1. Algorithm Complexity Profiler

import time

student_id = "0012"   # change this

# Main list (sorted for binary search)
main_list = [5, 12, 18, 23, 27, 31, 34, 45, 52, 61, 73, 84, 95]

# Small list with duplicate
small_list = [4, 7, 9, 2, 5, 7]

# Target = last 2 digits
target = int(student_id[-2:])

if target not in main_list:
    main_list.append(target)
    main_list.sort()

# Linear Search
start = time.time()
for num in main_list:
    if num == target:
        break
linear_time = time.time() - start

# Binary Search
start = time.time()
low, high = 0, len(main_list) - 1
while low <= high:
    mid = (low + high) // 2
    if main_list[mid] == target:
        break
    elif main_list[mid] < target:
        low = mid + 1
    else:
        high = mid - 1
binary_time = time.time() - start

# Quadratic duplicate check
start = time.time()
duplicate = None
for i in range(len(small_list)):
    for j in range(i + 1, len(small_list)):
        if small_list[i] == small_list[j]:
            duplicate = small_list[i]
            break
quadratic_time = time.time() - start

print("Algorithm\t\tTime (s)")
print("Linear Search\t\t", linear_time)
print("Binary Search\t\t", binary_time)
print("Quadratic Check\t\t", quadratic_time)
```

```python
# 2. Recursive vs Iterative Strategy Analysis

import time

student_id = "0012"

# n = last digit + 15
n = int(student_id[-1]) + 15

start = time.time()

# Iterative Fibonacci
a, b = 0, 1
if n == 0:
    fib = 0
elif n == 1:
    fib = 1
else:
    for _ in range(2, n + 1):
        a, b = b, a + b
    fib = b

elapsed = time.time() - start

print("n =", n)
print("Fibonacci =", fib)
print("Time =", elapsed)
```

```python
# 3. Sorting Algorithm Visual Comparator

import time
import random

student_id = "0012"

# Size = last 3 digits + 50
size = int(student_id[-3:]) + 50

data = random.sample(range(1, size * 10), size)


def bubble_sort(arr):
    comp, swaps = 0, 0
    start = time.time()
    for i in range(len(arr)):
        for j in range(len(arr) - i - 1):
            comp += 1
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swaps += 1
    return comp, swaps, time.time() - start


def selection_sort(arr):
    comp, swaps = 0, 0
    start = time.time()
    for i in range(len(arr)):
        min_idx = i
        for j in range(i + 1, len(arr)):
            comp += 1
            if arr[j] < arr[min_idx]:
                min_idx = j
        if min_idx != i:
            arr[i], arr[min_idx] = arr[min_idx], arr[i]
            swaps += 1
    return comp, swaps, time.time() - start


def insertion_sort(arr):
    comp, swaps = 0, 0
    start = time.time()
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0:
            comp += 1
            if arr[j] > key:
                arr[j + 1] = arr[j]
                swaps += 1
                j -= 1
            else:
                break
        arr[j + 1] = key
    return comp, swaps, time.time() - start


b = bubble_sort(data.copy())
s = selection_sort(data.copy())
i = insertion_sort(data.copy())

print("Algorithm\tComparisons\tSwaps\tTime")
print("Bubble\t\t", b)
print("Selection\t", s)
print("Insertion\t", i)
```
