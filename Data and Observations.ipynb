```python
# DATA AND OBSERVATION 2

numbers = [3, 7, 2, 9, 5]
numbers.sort()
target = 7
low = 0
high = len(numbers) - 1
while low <= high:
    mid = (low + high) // 2
    if numbers[mid] == target:
        print("Found at index", mid)
        break
    elif numbers[mid] < target:
        low = mid + 1
    else:
        high = mid - 1
```

```python
# DATA AND OBSERVATION 10

def fib_iter(n):
    a, b = 0, 1
    for _ in range(n):
        a, b = b, a + b
    return a
print(fib_iter(10))
```

```python
# DATA AND OBSERVATION 11

def fib_rec(n):
    if n <= 1:
        return n
    return fib_rec(n-1) + fib_rec(n-2)
print(fib_rec(10))
```

```python
# DATA AND OBSERVATION 12

import time
n = 20
start = time.time()
print(fib_iter(n))
end = time.time()
print("Execution time:", end - start)
```

```python
# DATA AND OBSERVATION 13

import random
arr = random.sample(range(1000), 20)
print("Unsorted:", arr)
```

```python
# DATA AND OBSERVATION 14

def bubble_sort(arr):
    n = len(arr)
    comparisons, swaps = 0, 0
    for i in range(n):
        for j in range(0, n-i-1):
            comparisons += 1
            if arr[j] > arr[j+1]:
                swaps += 1
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr, comparisons, swaps  
print(bubble_sort([5, 3, 8, 4, 2]))
```

```python
# DATA AND OBSERVATION 15

def selection_sort(arr):
    comparisons, swaps = 0, 0
    for i in range(len(arr)):
        min_idx = i
        for j in range(i+1, len(arr)):
            comparisons += 1
            if arr[j] < arr[min_idx]:
                min_idx = j
        swaps += 1
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr, comparisons, swaps
print(selection_sort([5, 3, 8, 4, 2]))
```

```python
# DATA AND OBSERVATION 16

def insertion_sort(arr):
    comparisons, swaps = 0, 0
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            comparisons += 1
            arr[j + 1] = arr[j]
            swaps += 1
            j -= 1
        arr[j + 1] = key
    return arr, comparisons, swaps
print(insertion_sort([5, 3, 8, 4, 2]))
```

```python
# DATA AND OBSERVATION 17

import time
arr = random.sample(range(1000), 100)
start = time.time()
bubble_sort(arr[:])
end = time.time()
print("Bubble sort time:", end - start)
```
